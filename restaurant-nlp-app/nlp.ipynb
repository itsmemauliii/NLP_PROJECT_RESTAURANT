```python
import pandas as pd
```

```python
df = pd.read_csv("/IndianFoodDatasetCSV.csv")
df
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
print(df["Cuisine"].value_counts())
```

```python
df["Ingredients_Filled"] = df["TranslatedIngredients"].fillna(df["Ingredients"])
df["Ingredients_Filled"] = df["Ingredients_Filled"].fillna("")
import re
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["Ingredients_Clean"] = df["Ingredients_Filled"].apply(clean_text)
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

if "Dish_Name" in df.columns:
    df["Recipe_Name_Clean"] = df["RecipeName"].apply(clean_text)

if "Ingredients" in df.columns:
    df["Ingredients_Clean"] = df["Ingredients"].apply(clean_text)

if "Cuisine" in df.columns:
    df["Cuisine_Clean"] = df["Cuisine"].astype(str).str.strip().str.title()

if "Course" in df.columns:
    df["Course_Clean"] = df["Course"].astype(str).str.strip().str.title()

# Any other text columns? Clean them too
for col in df.select_dtypes(include="object").columns:
    if not col.endswith("_Clean"):
        df[col + "_CleanedAuto"] = df[col].apply(clean_text)
```

```python
print("Missing values:\n", df.isnull().sum())
print("Missing in Ingredients:", df["Ingredients"].isnull().sum())
print("Missing in TranslatedIngredients:", df["TranslatedIngredients"].isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nUnique cuisines:\n", df["Cuisine_Clean"].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt

df["Cuisine_Clean"].value_counts().head(10).plot(kind="bar", figsize=(8,5))
plt.title("Top 10 Cuisines")
plt.show()
```

```python
from collections import Counter

all_ingredients = " ".join(df["Ingredients_Clean"]).split()
common_ing = Counter(all_ingredients).most_common(20)

ingredients, counts = zip(*common_ing)
plt.barh(ingredients[::-1], counts[::-1])
plt.title("Most Common Ingredients")
plt.show()
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=800, height=400, background_color="white").generate(" ".join(df["Recipe_Name_Clean"]))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
